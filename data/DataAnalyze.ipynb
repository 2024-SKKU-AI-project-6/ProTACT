```python
from transformers import BertTokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
# Function to check token lengths
def check_token_lengths(paragraphs, tokenizer, max_length=512):
    lengths = []
    for paragraph in paragraphs:
        tokens = tokenizer(paragraph, truncation=False)['input_ids']
        lengths.append(len(tokens))
    return lengths

# Check token lengths
token_lengths = check_token_lengths(paragraph, tokenizer)

# Find paragraphs that exceed the maximum token length
exceeding_paragraphs = [i for i, length in enumerate(token_lengths) if length > 512]

# Output the results
print(f"Token lengths: {token_lengths}")
if exceeding_paragraphs:
    print(f"Paragraphs exceeding 512 tokens: {exceeding_paragraphs}")
else:
    print("No paragraphs exceed the 512 token limit.")
```
